```python
plt.plot([1,2,3],[4,5,6],label='line1')
plt.plot([4,5,6],[6,41,5],label='line2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('test Graph')
plt.legend()# 用于生成默认图例
plt.show()
```

```python
x1=[1,2,3]
y1=[4,5,6]
z1=[7,8,9]
x2=[4,5,6]
y2=[6,41,5]
z2=[7,8,9]
plt.scatter(x1,y1,z1,label='dot1',marker='>')
plt.scatter(x2,y2,z2,marker='o',label='dot2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('this is a scatter')
plt.show()
```

```python
fig = plt.figure()
ax1=fig.add_subplot(1,1,1,projection='3d')
ax2=fig.add_subplot(2,2,2)
ax1.plot(x1,y1,z1)
ax2.scatter(x2,y2)
plt.show()
```
